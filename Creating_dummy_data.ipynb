```python
#  [Problem 1] Creating random numbers 
# ------------------------------------
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
mean = [-3, 0]
cov = [[1, 0.8],[0.8, 1]]
data = np.random.multivariate_normal(mean, cov, size=500)
# [Problem 2] Visualization by scatter plot
plt.figure("""first figure""")
plt.scatter(data[:,0], data[:,1])
plt.axis('equal')
plt.show()

# [Problem 3] Visualization by histogram
plt.figure("""second figure""")
plt.hist(data[:,0])
plt.figure("""third figure""")
plt.hist(data[:,1])

# [Problem 4] Addition of data
mean_add = [0, -3]
cov_add = [[1, 0.8],[0.8, 1]]
data_add = np.random.multivariate_normal(mean_add, cov_add, size=500)
plt.figure("""fourth figure""")
plt.scatter(data[:,0], data[:,1])
plt.scatter(data_add[:,0], data_add[:,1])
plt.axis('equal')
plt.show()

# Problem 5 Data combination
np.concatenate((data, data_add))


# Problem 6 Labeling
label0 = np.zeros((500,1))
label1 = np.ones((500,1))
data0 = np.concatenate((data, label0), axis=1)
data1 = np.concatenate((data_add, label1), axis=1)
data_new=np.concatenate((data0, data1), axis=0)
np.shape(data_new)
```
